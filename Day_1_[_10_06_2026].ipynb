```
import os
import datetime
import random

print("--- Intermediate Python Core Exercise ---")

# Global Data
DEVICE_TYPES = ['Temperature_Sensor', 'Humidity_Sensor', 'Pressure_Sensor', 'Light_Sensor']
LOCATIONS = ['Lab1', 'Lab2', 'OfficeA', 'OfficeB', 'Warehouse']

READING_RANGES = {
    'Temperature_Sensor': (18.0, 30.0),
    'Humidity_Sensor': (30.0, 80.0),
    'Pressure_Sensor': (980.0, 1030.0),
    'Light_Sensor': (100, 1000)
}

# --------------------------------------------------
# Task 1.1 Generate Sensor Data
# --------------------------------------------------

raw_sensor_data = []

for i in range(100):
    device_type = random.choice(DEVICE_TYPES)
    location = random.choice(LOCATIONS)

    min_val, max_val = READING_RANGES[device_type]
    value = round(random.uniform(min_val, max_val), 2)

    timestamp = (
        datetime.datetime.now() +
        datetime.timedelta(minutes=i)
    ).strftime('%Y-%m-%d %H:%M:%S')

    device_id = f"DEV{i+1:03d}"

    raw_sensor_data.append(
        (timestamp, device_id, device_type, location, value)
    )

# --------------------------------------------------
# Task 1.2 Temperature Readings
# --------------------------------------------------

temperature_readings = []

for reading in raw_sensor_data:
    if reading[2] == "Temperature_Sensor":
        temp_dict = {
            "timestamp": reading[0],
            "device_id": reading[1],
            "location": reading[3],
            "temperature_celsius": reading[4]
        }
        temperature_readings.append(temp_dict)

# --------------------------------------------------
# Task 1.3 Unique Locations
# --------------------------------------------------

unique_locations = set()

for reading in raw_sensor_data:
    unique_locations.add(reading[3])

print("\nUnique Locations:")
print(unique_locations)

# --------------------------------------------------
# Task 1.4 Device Stats
# --------------------------------------------------

device_stats = {}

for device in DEVICE_TYPES:
    values = []

    for reading in raw_sensor_data:
        if reading[2] == device:
            values.append(reading[4])

    if values:
        device_stats[device] = (min(values), max(values))

print("\nDevice Stats:")
print(device_stats)

# --------------------------------------------------
# Task 1.5 Advanced Indexing and Slicing
# --------------------------------------------------

sample_string = "PythonProgrammingIsFun"

sliced_readings_1 = raw_sensor_data[9:20]
sliced_readings_2 = raw_sensor_data[::5]

substring_1 = sample_string[6:17]
substring_2 = sample_string[0:6][::-1]

fifth_reading_device_type = raw_sensor_data[4][2]
device_type_suffix = fifth_reading_device_type[-6:]

print("\nSubstring 1:", substring_1)
print("Substring 2:", substring_2)
print("5th Device Type:", fifth_reading_device_type)
print("Suffix:", device_type_suffix)

# --------------------------------------------------
# Task 2.1 Function
# --------------------------------------------------

def get_readings_by_location(readings_list, target_location):
    result = []

    for reading in readings_list:
        if reading[3] == target_location:
            result.append(reading)

    return result

# --------------------------------------------------
# Task 2.2 Function
# --------------------------------------------------

def calculate_average_temperature(temp_data_list):

    if len(temp_data_list) == 0:
        return 0.0

    total = 0

    for item in temp_data_list:
        total += item["temperature_celsius"]

    return total / len(temp_data_list)

# --------------------------------------------------
# Task 2.3 Function
# --------------------------------------------------

def generate_report_line(reading_dict):
    return (
        f"[{reading_dict['timestamp']}] "
        f"Device {reading_dict['device_id']} "
        f"at {reading_dict['location']}: "
        f"{reading_dict['temperature_celsius']}°C"
    )

avg_temp = calculate_average_temperature(temperature_readings)

print("\nAverage Temperature:", round(avg_temp, 2))

# --------------------------------------------------
# Task 3.1 Create Directory
# --------------------------------------------------

reports_dir = "sensor_reports"

if not os.path.exists(reports_dir):
    os.makedirs(reports_dir)
    print("\nDirectory created.")
else:
    print("\nDirectory already exists.")

# --------------------------------------------------
# Task 3.2 Duration Calculation
# --------------------------------------------------

if len(raw_sensor_data) > 1:

    first_timestamp = raw_sensor_data[0][0]
    last_timestamp = raw_sensor_data[-1][0]

    format_str = "%Y-%m-%d %H:%M:%S"

    first_dt = datetime.datetime.strptime(
        first_timestamp,
        format_str
    )

    last_dt = datetime.datetime.strptime(
        last_timestamp,
        format_str
    )

    duration = last_dt - first_dt

    days = duration.days
    hours = duration.seconds // 3600
    minutes = (duration.seconds % 3600) // 60

    print(
        f"\nDuration: {days} days, "
        f"{hours} hours, {minutes} minutes"
    )

# --------------------------------------------------
# Task 3.3 Random Values
# --------------------------------------------------

random_values = []

for i in range(5):
    random_values.append(random.randint(100, 200))

print("\nRandom Values:")
print(random_values)

# --------------------------------------------------
# Task 4.1 Write Report File
# --------------------------------------------------

report_filename = os.path.join(
    reports_dir,
    "temperature_report.txt"
)

try:
    with open(report_filename, "w") as file:

        for reading in temperature_readings:
            line = generate_report_line(reading)
            file.write(line + "\n")

    print("\nTemperature report written successfully.")

except IOError as e:
    print("Error:", e)

# --------------------------------------------------
# Task 4.2 Read Report File
# --------------------------------------------------

read_report_lines = []

try:
    with open(report_filename, "r") as file:
        read_report_lines = file.readlines()

    print("\nFirst 3 Lines of Report:")

    for line in read_report_lines[:3]:
        print(line.strip())

except IOError as e:
    print("Error:", e)

print("\n--- Exercise Complete! ---")
```

```
import numpy as np

u = np.array([2, 3])
v = np.array([4, 1])

print("u + v =", u + v)      # [6, 4]
print("3 * u =", 3 * u)      # [6, 9]
```

```
#dot product
dot1 = np.dot(u, v)          # 2*4 + 3*1 = 11
dot2 = u @ v                 # Same result

cos_theta = np.dot(u, v) / (
    np.linalg.norm(u) * np.linalg.norm(v)
)

theta = np.arccos(cos_theta)

print(f"Angle: {np.degrees(theta):.1f}°")
```

**Matrix Operations on NumPy**

```
# 🛠 Hands-On #2: Matrix Operations in NumPy

## Matrix Creation & Basic Ops

import numpy as np

A = np.array([[2, 1], [0, 3]])
B = np.array([[1, 4], [2, 5]])

print("A + B =\n", A + B)
print("A transpose =\n", A.T)
print("A @ B =\n", A @ B)    # Matrix multiplication

## Identity & Inverse


I = np.eye(3)                # 3x3 identity matrix
print("Identity:\n", I)

A = np.array([[2, 1], [1, 3]])
A_inv = np.linalg.inv(A)

print("A inverse:\n", A_inv)
print("A @ A_inv =\n", A @ A_inv)   # Should be identity


## Solving Ax = b


# Solve: 2x + y = 5, x + 3y = 7

A = np.array([[2, 1], [1, 3]])
b = np.array([5, 7])

x = np.linalg.solve(A, b)    # More stable than inv(A) @ b

print("Solution:", x)        # [1.6, 1.8]

# Verify: A @ x should equal b
print("Check: A @ x =", A @ x)   # [5.0, 7.0]


```
